```python
sns.countplot(x='Survived',data=df)
plt.xlabel("Survival Status")
plt.ylabel("Number of Passengers")
plt.title("Passenger Survival Counts")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
sns.histplot(x="Age", hue="Survived", multiple="stack", data=df)
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.title("Passenger Age Distribution by Survival Status")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.boxplot(x='Pclass',y='Age',data=df,palette='crest')
```

```python
def input_age(cols):
    Age = cols[0]
    Pclass = cols[1]

    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
df['Age'] = df[['Age','Pclass']].apply(input_age,axis=1)
```

```python
#Dropped CAbin column due to many null values
df.head()
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='plasma')
```

```python
df.dropna(inplace=True)
```

```python
sex = pd.get_dummies(df['Sex'],drop_first=True)
embark = pd.get_dummies(df['Embarked'],drop_first=True)
df.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
df.head()
```

```python
df=pd.concat([df,sex,embark],axis=1)
df.head()
```
